```python
words =set()
content  = open('enable1.txt', 'r')
lines = content.readlines()
for line in lines:
    word= line.replace("b'", '').replace("\r", '').replace("\n", '')
    print(word)
    words.add(word)
```

```python
word_dict = dict()

for word in words:
    word_dict[word] = defaultdict(int)
    for char in word:
        word_dict[word][char] +=1
```

```python
import re

def dank(name, words):
    name = name.lower()
    name = re.sub("[^a-z]", "", name)

    mistakes = set()
    for word in words:
        workingname = name
        for char in word:
            if char in workingname:
                workingname = workingname.split(char)
                workingname = ''.join(workingname[1:])
                #print(newname)
            else:
                mistakes.add(word)
                break
    valids = words - mistakes
    
    maxlen = 0
    for item in valids:
        if len(item)>maxlen:
            maxlen = len(item)
    
    candidates = []
    for item in valids:
        if len(item) == maxlen:
            candidates.append(item)
    return(candidates)
```

```python
dank(name, words)
```

```python
for name in names:
    print(dank(name, words))
```
